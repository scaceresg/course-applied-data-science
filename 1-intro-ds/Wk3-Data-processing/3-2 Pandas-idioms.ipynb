## More Data Processing with Pandas

### Pandas Idioms

The best Python solutions to solve a particular problem are called **Idiomatic Python**.

A sort of sub-language within Python are Pandas Idioms. An example of these is using *vectorization* instead of iterative loops. Several developers and users within the Pandas community use the term **__ pandorable __** for these idioms.

```python
import pandas as pd
import numpy as np
import timeit
```

```python
# We will use the census data from US
df = pd.read_csv('../resources/week-3/datasets/census.csv')
df.head()
```

**Method Chaining**: 

In Method Chaining, every method on an object returns a reference to that object. You can condense many different operations on a DataFrame, for instance, into one line or at least one statement of code.

For example, let's pull out the state and city names as a multiple index, only for data which has a summary level of 50 (county-level data).

```python
# Example of method chaining (Pandorable)
(df.where(df['SUMLEV']==50) # Boolean mask for country level data
    .dropna() # .where() does not drop missing values by default
    .set_index(['STNAME', 'CTYNAME'])
    .rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})) # Rename to make it more readable
```

Let's measure how much time it's spent in both ways: Pandorable and Non-Pandorable

```python
# Pandorable approach
# First, we create a function for each approach
def pandorable_approach():

    global df

    return (df.where(df['SUMLEV']==50)
              .dropna()
              .set_index(['STNAME', 'CTYNAME'])
              .rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'}))

# Read a new dataset
df = pd.read_csv('../resources/week-3/datasets/census.csv')

# Then, we use timeit to run it 10 times and measure the time
timeit.timeit(pandorable_approach, number=10)
```

```python
# Non-Pandorable approach
def non_pandorable_approach():

    global df

    new_df = df[df['SUMLEV']==50]
    new_df.set_index(['STNAME', 'CTYNAME'], inplace=True)
    
    return new_df.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})

df = pd.read_csv('../resources/week-3/datasets/census.csv')

timeit.timeit(non_pandorable_approach, number=10)
```

The second approach is much faster! This is an example of time-readability trade off.

**`apply()`**:

`map()` is a built-in function in Python that you use to apply some function you want to an iterable (e.g. a list). This results in a list of all of the evaluations of that function.

`applymap()` is similar for Pandas because you provide a function that should operate on each cell of the DataFrame, and the return set is itself a DataFrame. However, `applymap()` is rarely used. Instead, `apply()` is heavily used because it maps across all of the rows in the DataFrame.

```python
# Let's look at the census data
df.head(10)
```

In this DataFrame, we have five columns for population estimates (`'POESTIMATE201x'`) with each column corresponding with one year of estimates.

Let's create new columns for minimum or maximum values by using the `apply()` function.

```python
# First, let's create a function to find min and max values
def min_max(row):

    # Project the data that we are interested in
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    
    # Return min, max values as a Series
    return pd.Series({'min': np.min(data), 'max': np.max(data)})
```

`apply()` takes the function and the axis on which to operate as parameters. 

**Important**: We've talked about `axis=0` for rows, however, for `apply()`, axis is the parameter of the index to use. So, we need to apply it across all rows, which is applying on all columns. We pass `axis='columns`.

```python
# Now we need to call apply on the DataFrame
df.apply(min_max, axis='columns').head()
```

Now, let's add the columns to the original DataFrame

```python
def min_max(row):

    # Project the data that we are interested in
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    
    # Create new entry for max and min
    row['MAX'] = np.max(data)
    row['MIN'] = np.min(data)
    
    return row
```

```python
# Now we need to call apply on the DataFrame
df.apply(min_max, axis='columns')
```

**`apply() + lambda`**:

Typically, the `apply()` function is used with the `lambda` function, usually, to chain several apply calls with lambdas together to create a readable yet succint data manipulation script.

For example, let's calculate the max of the columns using this joint approach:

```python
rows = ['POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2013', 'POPESTIMATE2014', 'POPESTIMATE2015']
```

```python
# Now, we'll just apply this across the dataframe with a lambda
df.apply(lambda x: np.max(x[rows]), axis=1).head()
```

**Remember**: `lambda` is an unnamed function in Python which takes a single parameter (`x`, in this case), and returns the maximum over all columns associated with row `x`.

Let's try another example of `apply()`: Let's say we want to divide the states into four categories: Northeast, Midwest, South, and West.

We can write a customised function that returns the region based on the state and the information of the region from Wikipedia.

```python
def get_state_region(x):

    northeast = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire',
                 'Rhode Island', 'Vermont', 'New York', 'New Jersey', 'Pennsylvania']
    midwest = ['Illinois', 'Indiana', 'Michigan', 'Ohio', 'Wisconsin', 'Iowa',
               'Kansas', 'Minnesota', 'Missouri', 'Nebraska', 'North Dakota',
               'South Dakota']
    south = ['Delaware', 'Florida', 'Georgia', 'Maryland', 'North Carolina',
             'South Carolina', 'Virginia', 'District of Columbia', 'West Virginia',
             'Alabama', 'Kentucky', 'Mississippi', 'Tennessee', 'Arkansas',
             'Lousiana', 'Oklahoma', 'Texas']
    west = ['Arizona', 'Colorado', 'Idaho', 'Montana', 'Nevada', 'New Mexico', 'Utah',
            'Wyoming', 'Alaska', 'California', 'Hawaii', 'Oregon', 'Washington']
    
    if x in northeast:
        return 'Northeast'
    elif x in midwest:
        return 'Midwest'
    elif x in south:
        return 'South'
    else:
        return 'West'
```

```python
# Now, let's create the new column using the apply function
df['state_region'] = df['STNAME'].apply(lambda x: get_state_region(x))
```
